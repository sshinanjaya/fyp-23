### Handle null values

#### Remove empty responses and empty columns

```python
df = df.dropna(how='all')
clean = df.dropna(axis=1, how='all')
clean = clean.reset_index(drop=['index'])
```

#### Replace null values with 0 or empty string accordingly

```python
clean[['age', 'total_child', 'f_child', 'm_child']] = clean[['age', 'total_child', 'f_child', 'm_child']].fillna(0)
```

```python
clean = clean.fillna('')
```

#### Change field types accordingly

#### Change to int

```python
clean[['age', 'total_child', 'f_child', 'm_child']] = clean[['age', 'total_child', 'f_child', 'm_child']].apply(pd.to_numeric, errors='coerce')
```

#### Change multi-valued fields from string to list of values

```python
multivalued_fields = ['f_child_age', 'm_child_age', 'reasons_eat_out', 'reasons_eat_in', 'reasons_fast_food', 'preference_cooking_method', 'reasons_preference_cooking_method',
       'who_make_decisions', 'type_grocery', 'reasons_child_influence',
       'reasons_no_child_influence']
```

```python
for field in multivalued_fields:
    clean[field] = clean[field].apply(lambda x: x.split(','))
```

```python
clean.head()
```

#### Change f_child_age and m_child_age from str to int obj

```python
for i in range (0, clean.shape[0]-1):

    # Row you want to remove items from
    row_index = i

    # List of column names you want to process
    columns_to_process = ['f_child_age', 'm_child_age']

    # Loop through the specified columns
    for column in columns_to_process:
        # Get the list at the specified row and column
        item_list = clean.at[row_index, column]
        
        # Remove trailing empty space
        item_list = [item.strip() for item in item_list]
        
        # Convert string to int
        item_list = [int(item) if item.isdigit() else item for item in item_list]
        
        # Update the DataFrame with the filtered list
        clean.at[row_index, column] = item_list
                
```

#### Remove ' please specify:' from list of obj

```python
for i in range (0, clean.shape[0]-1):

    # Row you want to remove items from
    row_index = i

    # List of column names you want to process
    columns_to_process = ['f_child_age', 'm_child_age', 'reasons_eat_out', 'reasons_eat_in', 'reasons_fast_food', 'preference_cooking_method', 'reasons_preference_cooking_method', 'who_make_decisions', 'type_grocery', 'reasons_child_influence', 'reasons_no_child_influence']

    # Loop through the specified columns
    for column in columns_to_process:
        # Get the list at the specified row and column
        item_list = clean.at[row_index, column]
        
        # Remove items equal to 'please specify:'
        filtered_list = [item for item in item_list if item != ' please specify:']
        
        # Update the DataFrame with the filtered list
        clean.at[row_index, column] = filtered_list
```

#### Remove unnecessary objects from type_grocery

```python
for i in range (0, clean.shape[0]-1):

    # Row you want to remove items from
    row_index = i

    filtered_list = clean.at[row_index, 'type_grocery']
    
    # Remove items equal to 'please specify:'
    filtered_list = [item for item in filtered_list if item != ' bread']
    filtered_list = [item for item in filtered_list if item != ' pasta)']
    filtered_list = [item for item in filtered_list if item != ' poultry']
    filtered_list = [item for item in filtered_list if item != ' fish']
    filtered_list = [item for item in filtered_list if item != ' beans']
    filtered_list = [item for item in filtered_list if item != ' tofu)']
    filtered_list = ['Proteins' if item == 'Proteins (e.g. meat' else item for item in filtered_list]
    filtered_list = ['Grains' if item == 'Grains (e.g. rice' else item for item in filtered_list]
    
    # Update the DataFrame with the filtered list
    clean.at[row_index, 'type_grocery'] = filtered_list
```

#### Change age object into integer

```python
for i in range (0, clean.shape[0]-1):

    # Row you want to remove items from
    row_index = i

    # List of column names you want to process
    columns_to_process = ['f_child_age', 'm_child_age']

    # Loop through the specified columns
    for column in columns_to_process:
        # Get the list at the specified row and column
        item_list = clean.at[row_index, column]
        
        # Remove items equal to 'please specify:'
        filtered_list = [item for item in item_list if item != ' please specify:']
        
        # Update the DataFrame with the filtered list
        clean.at[row_index, column] = filtered_list
```

### Remove descriptors for ease of analysis

```python
for i in range (0, clean.shape[0]-1):

    # Row you want to remove items from
    row_index = i

    # List of column names you want to process
    columns_to_process = ['child_influence_more_healthy_food', 'child_influence_new_recipe',
       'child_influence_grocery', 'include_child_in_meal_planning',
       'child_influence_less_unhealthy_food', 'prepare_separate_meals',
       'child_influence_learning_healthy_diet',
       'child_influence_overall_healthier_diet']

    # Loop through the specified columns
    for column in columns_to_process:
        # Get original string
        original_string = clean.at[row_index, column]
        
        # Find the position of the specific character in the string
        position = original_string.find(' - ')
        
        if position != -1:
            # Remove characters after the specific character
            result_string = original_string[:position]

            clean.at[row_index, column] = result_string
```
